# 📊 Flu Prediction - Demo Notebook
Dieses Notebook zeigt eine kleine **Datenanalyse für Influenza-Fälle**

**Funktionen:**
- Daten laden und anzeigen
- Einfache Datenanalyse
- Visualisierung von Trends


```python
# 📥 Bibliotheken importieren
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 📂 Daten laden (simulierte Beispiel-Daten)
data = pd.read_csv('flu_data.csv', parse_dates=['Datum'])

# 🔍 Erste 5 Zeilen anzeigen
data.head()
```

```python
# 📊 Basis-Datenanalyse
print('📌 Dateninfo:')
print(data.info())
print('\n📌 Beschreibende Statistik:')
print(data.describe())
```

```python
# 📈 Daten visualisieren
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Datum', y='Influenza_Fälle', label='Influenza-Fälle')
plt.title('📊 Influenza-Fälle über Zeit')
plt.xlabel('Datum')
plt.ylabel('Fälle')
plt.legend()
plt.show()
```
